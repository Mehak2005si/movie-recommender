```python
#imports
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
# keep the colms that are actually needed
#genre, id, keywords, title, overview, cast, crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()  #to find missing data
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#convert json like doc to readable text
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#get top 3 actors
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
#get directors
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
['love', 'loving', 'loved']
ps.stem('loving')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
user_movie_matrix = ratings.pivot_table(index='userId', columns='movieId', values='rating')
user_movie_matrix.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_movie_matrix)
```

```python
def get_collab_score(user_id, movie_id):
    if movie_id not in user_movie_matrix.columns:
        return 0
    
    similar_users = user_similarity[user_id]
    movie_ratings = user_movie_matrix[movie_id]
    
    score = np.dot(similar_users, movie_ratings) / (np.sum(similar_users) + 1e-8)
    return score / 5  # normalize
```

```python
def hybrid_recommend(movie, user_id, alpha=0.6):
    
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)), 
                         reverse=True, 
                         key=lambda x: x[1])[1:10]
    
    final_results = []
    
    for i in movies_list:
        movie_id = new_df.iloc[i[0]].movie_id
        
        content_score = i[1]
        collab_score = get_collab_score(user_id, movie_id)
        
        final_score = alpha * content_score + (1 - alpha) * collab_score
        
        final_results.append((new_df.iloc[i[0]].title, final_score))
    
    final_results = sorted(final_results, key=lambda x: x[1], reverse=True)
    
    for movie in final_results[:5]:
        print(movie[0])
```

```python
hybrid_recommend('Avatar', user_id=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(new_df['tags']).toarray()
```

```python
def recommend_from_text(user_input):
    user_vec = tfidf.transform([user_input]).toarray()
    
    sim_scores = cosine_similarity(user_vec, vectors)[0]
    
    movies_list = sorted(list(enumerate(sim_scores)), 
                         reverse=True, 
                         key=lambda x: x[1])[0:5]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
def chatbot():
    print("🎬 MovieBot: Tell me what you feel like watching!")
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == 'exit':
            break
        
        print("\nRecommended movies:")
        recommend_from_text(user_input)
        print("\n")
```

```python
def hybrid_chatbot(user_id):
    print("MovieBot")
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == 'exit':
            break
        
        user_vec = tfidf.transform([user_input]).toarray()
        sim_scores = cosine_similarity(user_vec, vectors)[0]
        
        top_movies = sorted(list(enumerate(sim_scores)), 
                            reverse=True, 
                            key=lambda x: x[1])[0:10]
        
        results = []
        
        for i in top_movies:
            movie_id = new_df.iloc[i[0]].movie_id
            
            content_score = i[1]
            collab_score = get_collab_score(user_id, movie_id)
            
            final_score = 0.6 * content_score + 0.4 * collab_score
            
            results.append((new_df.iloc[i[0]].title, final_score))
        
        results = sorted(results, key=lambda x: x[1], reverse=True)
        
        print("\n Recommended:")
        for movie in results[:5]:
            print(movie[0])
        print("\n")
```

```python
hybrid_chatbot(user_id = 1)
```
